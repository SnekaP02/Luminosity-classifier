```python
import pandas as pd
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
df_copy = df.copy()
```

```python
def clean_text(text):

    text = text.lower()
    
    text = re.sub(r'[^\w\s]', '', text)
    
    tokens = word_tokenize(text)
    
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(word) for word in tokens]
    
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    cleaned_text = ' '.join(tokens)
    
    return cleaned_text 
```

```python
# Apply the clean_text function to the 'Extracted_Content' column
df_copy['Cleaned_Content'] = df_copy['Extracted_Content'].apply(clean_text)

df_copy['Cleaned_Content'] = df_copy['Cleaned_Content'].apply(lambda x: re.sub(r'<[^>]+>', '', x))

df_copy.to_csv("/Users/sneka/Downloads/cleaned_train_dataset.csv", index=False) 
```

```python
def print_content(index):
    example = df_copy[df_copy.index == index][['Extracted_Content', 'Is lighting product?']].values[0]
    if len(example) > 0:
        print(example[0])
        print('Is lighting product?', example[1]) 
print_content(45)
```

```python
lighting_keywords = ["LED", "light", "illumination", "lumens", "lighting", "bulb"]

def is_lighting_product(text):
    for keyword in lighting_keywords:
        if keyword.lower() in text.lower():
            return "Yes"  
    return "No"  

# Apply the function to the 'Extracted_Content' column to classify each product
cleaned_data['Predicted_Is_Lighting_Product'] = cleaned_data['Extracted_Content'].apply(is_lighting_product)

# Evaluate the performance (compare 'Is lighting product?' column with 'Predicted_Is_Lighting_Product' column)
accuracy = (cleaned_data['Is lighting product?'] == cleaned_data['Predicted_Is_Lighting_Product']).mean()
print("Accuracy:", accuracy)
```
